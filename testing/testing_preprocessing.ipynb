## 1.1 Handling Missing Values

It can be seen that only 'MINIMUM_PAYMENTS' and 'CREDIT_LIMIT' have missing values. Hence, they are imputed with the following approaches

### 1.1.1: Imputing 'MINIMUM_PAYMENTS' with 'PAYMENTS', 'CREDIT_LIMIT' with median

The possible minimum payment made is at most the total payment made by the cardholder.
Therefore, the missing values in 'MINIMUM_PAYMENTS' are imputed with the corresponding 'PAYMENTS' made by the cardholder,
whereas the missing values in 'CREDIT_LIMIT' are imputed with the median of the 'CREDIT_LIMIT'

```python
# Creating a deep copy of the raw dataset
df_imputed_payments = df_raw.copy(deep = True)

# Imputing with the aforementioned method and storing in new columns
df_imputed_payments['MINIMUM_IMPUTED_MINIMUM_PAYMENTS'] = df_imputed_payments['MINIMUM_PAYMENTS'].fillna(df_imputed_payments['PAYMENTS'])
df_imputed_payments['IMPUTED_CREDIT_LIMIT'] = df_imputed_payments['CREDIT_LIMIT'].fillna(df_imputed_payments['CREDIT_LIMIT'].median())

# Saving to csv for future use
df_imputed_payments.to_csv(r'../raw_data/customer_imputed_payments.csv')

df_imputed_payments
```

## 2.0 Detecting Outliers

Both interquartile range (IQR) and z-Score methods are used to detect outliers in the dataset, but IQR is more suitable as the data are skewed, not near-normal, as indicated by its skewness and excess kurtosis

### 2.0.1 IQR Method
This method requires the first and third quartile of the dataset to be computed, which in turn is used to calculate the interquartile range, and in turn used to derive the upper and lower fences. Data points falling out of these fences are considered outliers. The equations are as follows:
\begin{align*}
Q1 &= \text{First Quartile (25th percentile)} \\
Q3 &= \text{Third Quartile (75th percentile)} \\
IQR &= Q3 - Q1 \\
\text{Lower Fence} &= Q1 - 1.5 \times IQR \\
\text{Upper Fence} &= Q3 + 1.5 \times IQR \\
\text{Outliers} &= \{ x \mid x < \text{Lower Fence} \text{ or } x > \text{Upper Fence} \}
\end{align*}

```python
# Choosing the KNN imputed dataset for further process and analysis
df_imputed = pd.read_csv(r'../raw_data/customer_imputed_knn_n3.csv', index_col = 'CUST_ID')

# Defining function for computing outliers using IQR method
def outliers_iqr(column):
    # Getting first and third quartile of the column
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    
    # Computing the IQR, lower and upper fences
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    
    # Getting the outliers
    outliers = column[(column < lower_fence) | (column > upper_fence)]
    
    # Returning the number of outliers, lower fence, upper fence, and percentage of outliers
    return len(outliers), lower_fence, upper_fence, round(len(outliers) / len(column) * 100, 2)

# Getting outliers information using IQR method
df_outliers_iqr = df_imputed.apply(outliers_iqr).T.rename({0: 'Outliers Count', 1: 'Lower Fence', 2: 'Upper Fence', 3: 'Percentage Outliers (%)'}, axis = 1)

df_outliers_iqr
```

## 2.0.2 z-Score Method
This method assumes an approximate normal distribution of the data, which is not the case for this dataset. Hence, it is not practical for outliers detection in this dataset. The z-Score is calculated as follows:
\begin{align*}
z &= \frac{x - \mu}{\sigma} \\
\text{Lower Fence} &= \mu - 3 \times \sigma \\
\text{Upper Fence} &= \mu + 3 \times \sigma \\
\text{Outliers} &= \{ x \mid z < -3 \text{ or } z > 3 \}
\end{align*}

```python
def outliers_zscore(column):
    # Computing the mean and standard deviation of the column
    mean = column.mean()
    std = column.std()
    
    # Computing the z-scores of each row
    z_scores = (column - mean) / std
    
    # Computing the lower and upper fences
    lower_fence = mean - 3 * std
    upper_fence = mean + 3 * std
    
    # Getting the outliers
    outliers = column[abs(z_scores) > 3]
    
    # Returning the number of outliers, lower fence, upper fence, and percentage of outliers
    return len(outliers), lower_fence, upper_fence, round(len(outliers) / len(column) * 100, 2)

# Getting outliers information using z-Score method
df_outliers_zscore = df_imputed.apply(outliers_zscore).T.rename({0: 'Outliers Count', 1: 'Lower Fence', 2: 'Upper Fence', 3: 'Percentage Outliers (%)'}, axis = 1)

df_outliers_zscore
```
