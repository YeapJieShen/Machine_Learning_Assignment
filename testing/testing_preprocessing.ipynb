## 2.1 Handling Missing Values
It shows that MINIMUM_PAYMENTS, CREDIT_LIMIT has missing values only

#### 2.1.1.1: Imputing MINIMUM_PAYMENTS with PAYMENTS, CREDIT_LIMIT with median
It is possible that the minimum payment made is at most the total payment made by the customer 

```python
pd.concat([df_raw['PAYMENTS'], df_raw['MINIMUM_PAYMENTS'].fillna(df_raw['PAYMENTS'])], axis = 1, keys = ['PAYMENTS', 'MINIMUM_IMPUTED_MINIMUM_PAYMENTS'])
```

```python
df_minimum_imputed_payments = df_raw.copy(deep = True)
df_minimum_imputed_payments['MINIMUM_IMPUTED_MINIMUM_PAYMENTS'] = df_minimum_imputed_payments['MINIMUM_PAYMENTS'].fillna(df_minimum_imputed_payments['PAYMENTS'])
df_minimum_imputed_payments['IMPUTED_CREDIT_LIMIT'] = df_minimum_imputed_payments['CREDIT_LIMIT'].fillna(df_minimum_imputed_payments['CREDIT_LIMIT'].median())
df_minimum_imputed_payments
```

#### 2.1.1.2 Imputing with Mean
Not practical since some customers may have made no payments at all, hence the minimum payment should not be the mean of the payments

```python
from sklearn.impute import SimpleImputer

imputer_mean = SimpleImputer(strategy = 'mean')

df_minimum_imputed_mean = df_raw.copy(deep = True)
df_minimum_imputed_mean = pd.DataFrame(imputer_mean.fit_transform(df_minimum_imputed_mean), columns = df_minimum_imputed_mean.columns, index = df_minimum_imputed_mean.index).rename({'MINIMUM_PAYMENTS': 'MEAN_IMPUTED_MINIMUM_PAYMENTS', 'CREDIT_LIMIT': 'MEAN_IMPUTED_CREDIT_LIMIT'}, axis = 1)
df_minimum_imputed_mean
```

### 2.1.1.3 Imputing with Median
Not practical as well, same reason as above

```python
imputer_median = SimpleImputer(strategy = 'median')

df_minimum_imputed_median = df_raw.copy(deep = True)
df_minimum_imputed_median = pd.DataFrame(imputer_median.fit_transform(df_minimum_imputed_median), columns = df_minimum_imputed_median.columns, index = df_minimum_imputed_median.index).rename({'MINIMUM_PAYMENTS': 'MEDIAN_IMPUTED_MINIMUM_PAYMENTS', 'CREDIT_LIMIT': 'MEDIAN_IMPUTED_CREDIT_LIMIT'}, axis = 1)
df_minimum_imputed_mean
```

### 2.1.1.4 Imputing with Zero
Not that practical, since some cardholders have made payments, hence the minimum payment should not be zero

```python
df_minimum_imputed_zero= df_raw.copy(deep = True)
df_minimum_imputed_zero = df_minimum_imputed_zero.fillna(0).rename({'MINIMUM_PAYMENTS': 'ZERO_IMPUTED_MINIMUM_PAYMENTS', 'CREDIT_LIMIT': 'ZERO_IMPUTED_CREDIT_LIMIT'}, axis = 1)
df_minimum_imputed_zero
```

### 2.1.1.5 Imputing with Forward Fill
Not practical at all, this is not a time series data

```python
df_minimum_imputed_ffill = df_raw.copy(deep = True)
df_minimum_imputed_ffill = df_minimum_imputed_ffill.ffill().rename({'MINIMUM_PAYMENTS': 'FFILL_IMPUTED_MINIMUM_PAYMENTS', 'CREDIT_LIMIT': 'FFILL_IMPUTED_CREDIT_LIMIT'}, axis = 1)
df_minimum_imputed_ffill
```

### 2.1.1.5 Imputing with Backward Fill
Not practical at all, this is not a time series data

```python
df_minimum_imputed_bfill = df_raw.copy(deep = True)
df_minimum_imputed_bfill = df_minimum_imputed_bfill.bfill().rename({'MINIMUM_PAYMENTS': 'BFILL_IMPUTED_MINIMUM_PAYMENTS', 'CREDIT_LIMIT': 'BFILL_IMPUTED_CREDIT_LIMIT'}, axis = 1)
df_minimum_imputed_bfill
```

### 2.1.1.6 Imputing with K-Nearest Neighbors Imputer
Good only for fitting into simple models, SimpleImputer would be considerably better for complex models

Reference:
https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html#sklearn.impute.SimpleImputer

```python
from sklearn.impute import KNNImputer

imputer_knn = KNNImputer(n_neighbors = 3)

df_minimum_knn_imputed = df_raw.copy(deep = True)
df_minimum_knn_imputed = pd.DataFrame(imputer_knn.fit_transform(df_minimum_knn_imputed), columns = df_minimum_knn_imputed.columns, index = df_minimum_knn_imputed.index).rename({'MINIMUM_PAYMENTS': 'KNN_IMPUTED_MINIMUM_PAYMENTS', 'CREDIT_LIMIT': 'KNN_IMPUTED_CREDIT_LIMIT'}, axis = 1)
df_minimum_knn_imputed
```

#### 2.1.1.7 Imputing with Iterative Imputer
Good only for fitting into simple models, SimpleImputer would be considerably better for complex models. It is using BayesianRidge as the estimator by default

Reference: https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html#sklearn.impute.SimpleImputer

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer_iterative = IterativeImputer()
df_minimum_imputed_iter = df_raw.copy(deep = True)
df_minimum_imputed_iter = pd.DataFrame(imputer_iterative.fit_transform(df_minimum_imputed_iter), columns = df_minimum_imputed_iter.columns, index = df_minimum_imputed_iter.index).rename({'MINIMUM_PAYMENTS': 'ITER_IMPUTED_MINIMUM_PAYMENTS', 'CREDIT_LIMIT': 'ITER_IMPUTED_CREDIT_LIMIT'}, axis = 1)
df_minimum_imputed_iter
```
